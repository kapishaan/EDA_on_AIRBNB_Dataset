```python
#convert neighbourhood into numeric data
neighbourhood = pd.DataFrame(airbnb['neighbourhood'])
```

```python
neighbourhood
```

```python
#sparse is not supported in 1.2 or later version instead of that we have to use sparse_output

encoder = OneHotEncoder(sparse_output=False) #Convert Categorical Data into Numeric Format
k = encoder.fit_transform(neighbourhood)
k
```

```python
encoder.categories_
```

```python
p = pd.DataFrame(k,columns = encoder.categories_)
```

```python
p
```

```python
new_airbnb = pd.concat([airbnb,p], axis=1)
```

```python
new_airbnb
```
